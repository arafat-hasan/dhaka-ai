```
#!/usr/bin/env python

import os
import random
import glob
from tqdm import tqdm

trainval_percent = 1.0
train_percent = 0.85


xmlfilepath = '/gdrive/My Drive/dhaka_ai/IDD-Dataset/IDD_Detection_25k/Annotations'
txtsavepath = '/gdrive/My Drive/dhaka_ai/IDD-Dataset/IDD_Detection_25k/ImageSets/Main/'

total_xml = glob.glob(os.path.join(xmlfilepath, '*.xml'))

num = len(total_xml)
list = range(num)
tv = int(num*trainval_percent)
tr = int(tv*train_percent)
trainval = random.sample(list, tv)
train = random.sample(trainval, tr)
ftrainval = open(os.path.join(txtsavepath, 'trainval.txt'), 'w')
ftest = open(os.path.join(txtsavepath, 'test.txt'), 'w')
ftrain = open(os.path.join(txtsavepath, 'train.txt'), 'w')
fval = open(os.path.join(txtsavepath, 'val.txt'), 'w')

for i in tqdm(list):
    name = os.path.basename(total_xml[i])[:-4]+'\n'
    if i in trainval:
        ftrainval.write(name)
        if i in train:
            ftrain.write(name)
        else:
            fval.write(name)
    else:
        ftest.write(name)

ftrainval.close()
ftrain.close()
fval.close()
ftest.close()
```

```
#!/usr/bin/env python

import shutil
import os
import random
import glob
from tqdm import tqdm


labelfilepath = '/gdrive/My Drive/dhaka_ai/IDD-Dataset/IDD_Detection_25k/darknettxt'
imagefilepath = '/gdrive/My Drive/dhaka_ai/IDD-Dataset/IDD_Detection_25k/JPEGImages'

trainimagepath = '/gdrive/My Drive/dhaka_ai/IDD-Dataset/IDD_Detection_25k_yolo/images/train'
valimagepath = '/gdrive/My Drive/dhaka_ai/IDD-Dataset/IDD_Detection_25k_yolo/images/val'
trainlabelpath = '/gdrive/My Drive/dhaka_ai/IDD-Dataset/IDD_Detection_25k_yolo/labels/train'
vallabelpath = '/gdrive/My Drive/dhaka_ai/IDD-Dataset/IDD_Detection_25k_yolo/labels/val'

all_total_file = glob.glob(os.path.join(imagefilepath, '*.jpg'))

copied_files_train = glob.glob(os.path.join(trainlabelpath, '*.txt'))
copied_files_val = glob.glob(os.path.join(vallabelpath, '*.txt'))
copied_files = copied_files_train + copied_files_val
copied_basenames = [os.path.splitext(os.path.basename(i))[0] for i in copied_files]
print(f'\nTotal already copied files count {len(copied_files)}')

train_percent = 0.85

total_train_file = len(all_total_file) * train_percent
total_val_file = len(all_total_file) - total_train_file

more_train_file = total_train_file - len(copied_files_train)
more_val_file = total_train_file - len(copied_files_val)

new_train_percent = more_train_file / (more_train_file + more_val_file)
print(f'New train percent: {new_train_percent}')
print("Separating copied files")
total_file = []
naked_file_names = []
for file in tqdm(all_total_file):
  basename = os.path.splitext(os.path.basename(file))[0]
  if basename in copied_basenames:
    continue
  total_file.append(file)
  naked_file_names.append(basename)

num = len(total_file)
list = range(num)
tr = int(num*new_train_percent)
train = random.sample(list, tr)


print("Coppying files")
for i in tqdm(list):
    image_file_name = naked_file_names[i] + ".jpg"
    txt_file_name =  naked_file_names[i] + ".txt"
    if  i in train:
        shutil.copy2(os.path.join(labelfilepath, txt_file_name), trainlabelpath)
        shutil.copy2(os.path.join(imagefilepath, image_file_name), trainimagepath)
    else:
        shutil.copy2(os.path.join(labelfilepath, txt_file_name), vallabelpath)
        shutil.copy2(os.path.join(imagefilepath, image_file_name), valimagepath)
        
        
print(f'\nProcesses {i} operations')
```
